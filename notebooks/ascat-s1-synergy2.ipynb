```python
paths = yaml.safe_load(Path("../paths.yml").read_text())
root: Path = Path(paths["linux"]).expanduser()
cell_source: Path = root / "datasets/scat_ard/ascat_ssm_cdr_12.5km_h121"
assert cell_source.exists()
```

```python
time_range = "2021-06-01/2021-09-30"

latmin, latmax = 47, 48  # South to North
lonmin, lonmax = 15, 16.5  # West to East

bounding_box = [lonmin, latmin, lonmax, latmax]
```

Let's extract `gpis` - grid point indices from read data.

```python
h121_reader = CellFileCollection.from_product_id(cell_source, "H121_V1.0")
gpis, lons, lats, cells = h121_reader.grid.get_grid_points()
```

Now, we will use specified bounding box that was used to filter out EODC's Sentinel-1 data to filter out grid point indices that are inside this area.

```python
indices = np.where(
    (lats >= latmin) & (lats <= latmax) & (lons >= lonmin) & (lons <= lonmax)
)[0]

print("There are", len(indices), "grid point indices inside specified bounding box")
```

Let's make an array of those indices that will be used to load individual time series of each point into an `xarray` dataarrays. Those dataarrays will be combined into `xarray` dataset, which will store time series of each individual grid point.

```python
selected_coords = list(zip(lons[indices], lats[indices]))
selected_gpis = gpis[indices]
```

```python
datasets = []

for i, id in enumerate(selected_gpis):
    print(f"Loading location_id: {id} ({i + 1}/{len(selected_gpis)})")

    ds = h121_reader.read(location_id=id)

    ds = ds.swap_dims({"obs": "time"})

    ds = ds.expand_dims(location_id=[id])
    datasets.append(ds)  # loop takes around 15 seconds

ascat_ds = xr.concat(datasets, dim="location_id")  # this will also take a few seconds
```

```python
ascat_ds = ascat_ds.sel(time=slice(*time_range.split("/")))
ascat_ds
```

```python
lats = ascat_ds["lat"].values
lons = ascat_ds["lon"].values
ids = ascat_ds["location_id"].values

center = [lats.mean(), lons.mean()]
map = folium.Map(location=center, zoom_start=8)

folium.Rectangle(
    bounds=[[latmin, lonmin], [latmax, lonmax]],
    color="blue",
    fill=True,
    fill_opacity=0.1,
    weight=2,
    popup="Area of Interest",
).add_to(map)

for lat, lon, loc_id in zip(lats, lons, ids):
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        popup=f"ID: {loc_id}",
        color="blue",
        fill=True,
        fill_color="blue",
    ).add_to(map)

map
```

### Loading Sentinel - 1 Sigma Naught Data from EODC STAC Catalogue

```python
eodc_catalog = pystac_client.Client.open("https://stac.eodc.eu/api/v1")

colllection_id = "SENTINEL1_SIG0_20M"

collection = eodc_catalog.get_collection(colllection_id)
```

Loading the metadata with STAC search engine.

```python
search = eodc_catalog.search(
    collections=colllection_id,
    bbox=bounding_box,
    datetime=time_range,
    # max_items=1  # number of max items to load
)
items_eodc = search.item_collection()
print(f"On EODC we found {len(items_eodc)} items for the given search query")
```

Let's plot thumbnail of the loaded items for this area and those dates.

```python
map = folium.Map(
    location=[(latmin + latmax) / 2, (lonmin + lonmax) / 2],
    zoom_start=7,
    zoom_control=False,
    scrollWheelZoom=False,
    dragging=False,
)

folium.Rectangle(
    bounds=[[latmin, lonmin], [latmax, lonmax]],
    color="blue",
    fill=True,
    fill_opacity=0.1,
    weight=2,
    popup="Area of Interest",
).add_to(map)

for item in items_eodc:
    # url leading to display of an item, can also be used as hyperlink
    image_url = item.assets["thumbnail"].href
    bounds = item.bbox
    folium.raster_layers.ImageOverlay(
        image=image_url,
        bounds=[[bounds[1], bounds[0]], [bounds[3], bounds[2]]],
    ).add_to(map)

folium.LayerControl().add_to(map)

map
```

Now, let's load the data with `odc_stac`.

```python
crs = "EPSG:4326"  # Coordinate Reference System: EQUI7 Grid of Europe
res = 0.0002  # meter
chunks = {"time": 1, "latitude": 1000, "longitude": 1000}
sig0_dc = odc_stac.load(
    items_eodc,
    crs=crs,
    resolution=res,
    bbox=bounding_box,
    chunks=chunks,
    resampling="bilinear",
)

sig0_dc
```

```python
item = items_eodc[0]
scale = item.assets["VV"].extra_fields.get("raster:bands")[0]["scale"]
nodata = item.assets["VV"].extra_fields.get("raster:bands")[0]["nodata"]
sig0_dc = sig0_dc.where(sig0_dc != nodata) / scale
sig0_dc
```

```python
sig0_dc.VV.isel(time=5)[::20, ::20].plot(cmap="viridis", robust=True)
```

## Sentinel data filtering per point grid

```python
buffer_deg = 0.001  # ~111 meters
sentinel_by_point = {}

total = ascat_ds.dims["location_id"]

for i in range(total):
    # print(f"Extracting {i + 1}/{total}...")

    lon0 = float(ascat_ds.lon[i].values)
    lat0 = float(ascat_ds.lat[i].values)

    window = {
        "longitude": slice(lon0 - buffer_deg, lon0 + buffer_deg),
        "latitude": slice(
            lat0 + buffer_deg, lat0 - buffer_deg
        ),
    }

    subset = sig0_dc.sel(window)
    sentinel_by_point[int(ascat_ds.location_id[i].values)] = subset
```

```python
vv_series = []
vh_series = []
times = None

for loc_id, ds in sentinel_by_point.items():
    try:
        # Sentinel lat/lon grid in the patch
        lats = ds.latitude.values
        lons = ds.longitude.values

        # ASCAT location
        lat0 = float(ascat_ds.lat.sel(location_id=loc_id))
        lon0 = float(ascat_ds.lon.sel(location_id=loc_id))

        # Compute distance to all pixels in patch
        dist = np.sqrt((lats[:, None] - lat0) ** 2 + (lons[None, :] - lon0) ** 2)
        i, j = np.unravel_index(np.argmin(dist), dist.shape)

        # Extract VV and VH from that pixel
        vv = ds.VV[:, i, j]
        vh = ds.VH[:, i, j]

        vv_series.append(vv)
        vh_series.append(vh)

        if times is None:
            times = ds.time.values

    except Exception as e:
        print(f"Skipping location {loc_id} due to: {e}")
        continue
```

```python
vv_da = xr.concat(vv_series, dim="location_id")
vh_da = xr.concat(vh_series, dim="location_id")

vv_da["location_id"] = ("location_id", ids)
vh_da["location_id"] = ("location_id", ids)

sig0_regridded_ds = xr.Dataset(
    {
        "VV": vv_da,
        "VH": vh_da,
    },
    coords={"location_id": ids, "time": times},
)

sig0_regridded_ds
```

```python
loc_id = 2

lat_s1 = sig0_regridded_ds.isel(location_id=loc_id).latitude.values
lon_s1 = sig0_regridded_ds.isel(location_id=loc_id).longitude.values

lat_ascat = float(ascat_ds.lat.isel(location_id=loc_id))
lon_ascat = float(ascat_ds.lon.isel(location_id=loc_id))

print(f"Sentinel-1 (nearest pixel) lat/lon: {lat_s1:.5f}, {lon_s1:.5f}")
print(f"ASCAT location lat/lon:              {lat_ascat:.5f}, {lon_ascat:.5f}")
```

```python
s1_vv_timeseries = sig0_regridded_ds.VV.isel(location_id=loc_id).dropna(dim="time")

ascat_sigma_timeseries = ascat_ds.backscatter40.isel(location_id=loc_id).dropna(dim="time")
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot Sentinel VV
ax1.plot(s1_vv_timeseries.time, s1_vv_timeseries, color="tab:blue", label="Sentinel-1 VV")
ax1.set_ylabel("VV [dB]", color="tab:blue")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Second axis for ASCAT
ax2 = ax1.twinx()
ax2.plot(ascat_sigma_timeseries.time, ascat_sigma_timeseries, color="tab:orange", label="ASCAT SM")
ax2.set_ylabel("ASCAT Backscatter [dB]", color="tab:orange")
ax2.tick_params(axis="y", labelcolor="tab:orange")

ax1.set_title(f"Location ID {loc_id} — S1 vs ASCAT time series")
ax1.set_xlabel("Time")

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper right")

plt.tight_layout()
plt.show()
```

### Temporal Matching

```python
# Get time arrays
s1_time = pd.to_datetime(s1_vv_timeseries["time"].values)
ascat_time = pd.to_datetime(ascat_sigma_timeseries["time"].values)

# Count per day
s1_counts = pd.Series(1, index=s1_time).resample("1D").sum()
ascat_counts = pd.Series(1, index=ascat_time).resample("1D").sum()

# Summary stats
s1_total = len(s1_time)
ascat_total = len(ascat_time)

s1_days = (s1_time.max() - s1_time.min()).days
ascat_days = (ascat_time.max() - ascat_time.min()).days

s1_rate = s1_total / s1_days
ascat_rate = ascat_total / ascat_days

s1_interval = s1_days / s1_total
ascat_interval = ascat_days / ascat_total


print(
    f"Sentinel-1 VV: {s1_total} obs in {s1_days} days — {s1_rate:.2f}/day (every {s1_interval:.2f} days)"
)
print(
    f"ASCAT SM:     {ascat_total} obs in {ascat_days} days — {ascat_rate:.2f}/day (every {ascat_interval:.2f} days)"
)

plt.figure(figsize=(12, 5))
ascat_counts.plot(label="ASCAT SM", alpha=0.8)
s1_counts.plot(label="Sentinel-1 VV", alpha=0.8)
plt.ylabel("Observations per day")
plt.title(f"Observation Frequency — Location Index {loc_id}")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
